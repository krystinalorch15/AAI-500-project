```python
#Importing required libs 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the healt data set using pandas lib 
healt_data = pd.read_csv ("healthcare_dataset.csv", sep =",")

# Remove all the duplicates from our data set 
healt_data = healt_data.drop_duplicates()

# Remove rows with null value 
healt_data = healt_data.dropna()

# Length of our data set 
print(f"Length of our data set is {len(healt_data)}")

# Check if there are any null values in the dataset after cleaning

has_nan = healt_data.isna().any().any()

if has_nan:
    print("There are null values in the Dataset.")
else:
    print("There are no null values in the Dataset.")
```

```python
# Plot of pariticpants by gender in our dataset 
sns.countplot(x=healt_data['Gender'], data=healt_data)
plt.title('Participants by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of participant age in our dataset
plt.figure(figsize=(8, 6))
plt.hist(healt_data['Age'], bins=50, edgecolor='black')  # Adjust number of bins as needed
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Group by 'gender' and 'blood_type'
grouped = healt_data.groupby(['Gender', 'Blood Type']).size().unstack()

# Plotting the grouped bar plot
grouped.plot(kind='bar', figsize=(10, 6))
plt.title('Blood Type Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.legend(title='Blood Type')
plt.show()
```

```python
# Group by 'gender' and 'Medical Condition'
grouped = healt_data.groupby(['Gender', 'Medical Condition']).size().unstack()

# Plotting the grouped bar plot
grouped.plot(kind='bar', figsize=(10, 6))
plt.title('Medical Condition Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.legend(title='Blood Type')
plt.show()
```

```python
# Here we define age groups
age_bins = [0, 20, 40, 60, 80, 100] 

# Lables for the above age groups 
age_labels = ['0-20', '21-40', '41-60', '61-80', '81-100']

# Categorize ages into groups 
healt_data['Age Group'] = pd.cut(healt_data['Age'], bins=age_bins, labels=age_labels, right=False)

# Group by 'Age Group' and 'Medical Condition', then count occurrences
grouped = healt_data.groupby(['Age Group', 'Medical Condition'], observed=False).size().unstack().fillna(0)

# Plotting the stacked bar plot
plt.figure(figsize=(12, 8))
grouped.plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Medical Condition Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Medical Condition', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
